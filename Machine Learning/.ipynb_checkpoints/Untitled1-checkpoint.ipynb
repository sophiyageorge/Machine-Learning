```python
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
X,y = load_breast_cancer(return_X_y=True)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.datasets import load_iris
import pandas as pd
```

```python
data = load_iris()
X,y =pd.DataFrame(data.data,columns=data.feature_names),data.target
X_train,X_test,y_train,y_test = train_test_split(X,y)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print(mean_squared_error(y_test,y_pred))
```

```python
print(r2_score(y_test,y_pred))
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Generate sample data
X, y = make_blobs(n_samples=300, centers=5, random_state=42)

# Elbow Method
wcss = []
k_values = range(1, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)  # Sum of squared distances to nearest cluster center

# Plot Elbow Curve
plt.plot(k_values, wcss, marker='o')
plt.title("Elbow Method")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("WCSS (Inertia)")
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

X,_ = make_blobs(n_samples=300,centers=5,cluster_std=0.6)
model = DBSCAN(eps=0.3,min_samples=5)
y = model.fit_predict(X)
print(silhouette_score(X,y))
plt.scatter(X[:,0],X[:,1],c=y,cmap='viridis')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

model = AgglomerativeClustering()
y= model.fit_predict(X)
print(silhouette_score(X,y))
plt.scatter(X[:,0],X[:,1],c=y,cmap='viridis')
plt.show()
```

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

X,_ = make_blobs(n_samples=300,centers=5,cluster_std=0.6,random_state=0)
model = KMeans(n_clusters=3)
y=model.fit_predict(X)
print(silhouette_score(X,y))
plt.scatter(X[:,0],X[:,1],c=y,cmap='viridis')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
data = load_iris()
X,y = data.data,data.target
scaler=StandardScaler()
X=scaler.fit_transform(X)

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
model = KNeighborsClassifier(n_neighbors=3,metric='euclidean')
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report
X=data.data[:,:2]
y=data.target
X=X[y!=2]
y=y[y!=2]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
model = SVC(kernel='linear',C=1.0)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier

data = load_iris()
X=data.data
y=data.target

X_train,X_test,y_train,y_test = train_test_split(X,y)
model = DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print(model.score(X_test,y_test))
```
